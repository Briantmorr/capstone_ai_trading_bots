```python
def compute_macd(data, fast=12, slow=26, signal=9):
    data['ema_fast'] = data['close'].ewm(span=fast, adjust=False).mean()
    data['ema_slow'] = data['close'].ewm(span=slow, adjust=False).mean()
    data['macd'] = data['ema_fast'] - data['ema_slow']
    data['signal'] = data['macd'].ewm(span=signal, adjust=False).mean()
    data['histogram'] = data['macd'] - data['signal']
    return data

df = compute_macd(df)
```

```python
symbols = ["AAPL", "GOOGL", "AMZN", "META", "MSFT", "NVDA"]
start_date = dt.datetime.now() - dt.timedelta(days=365)
end_date = dt.datetime.now()

request_params = StockBarsRequest(
    symbol_or_symbols=symbols,
    timeframe=TimeFrame.Minute,
    start=start_date,
    end=end_date
)

bars = client.get_stock_bars(request_params).df
stock_dic = {}
for symbol in symbols:
    df = bars[bars.index.get_level_values(0) == symbol].copy()
    df.reset_index(inplace=True)
    df['timestamp'] = pd.to_datetime(df['timestamp'])
    df.set_index('timestamp', inplace=True)
    df=compute_macd(df)
    stock_dic[symbol] = df

print(stock_dic["AAPL"].head())
print(stock_dic["GOOGL"].head())
print(stock_dic["AMZN"].head())
print(stock_dic["META"].head())
print(stock_dic["MSFT"].head())
print(stock_dic["NVDA"].head())
```

```python
initial_capital = 100_000
position_size_pct = 0.10
stop_loss_pct = 0.02
leverage = 2.0
trailing_stop_pct = 0.015
risk_free_rate = 0.01

capital = initial_capital
positions = []
in_position = False
entry_price = 0
position_direction = 0
trailing_stop = None
entry_time = None

equity_curve = []
returns = []
trade_log = []

for i in range(1, len(df)):
    row = df.iloc[i]
    prev_row = df.iloc[i - 1]

    if in_position:
        price_change = (row['close'] - entry_price) / entry_price * position_direction
        leveraged_return = leverage * price_change * position_size_pct

        if position_direction == 1:
            trailing_stop = max(trailing_stop, row['close'] * (1 - trailing_stop_pct))
            if row['close'] <= trailing_stop or price_change <= -stop_loss_pct:
                exit_time = row.name
                exit_price = row['close']
                trade_log.append({
                    'entry_time': entry_time,
                    'entry_price': entry_price,
                    'exit_time': exit_time,
                    'exit_price': exit_price,
                    'direction': 'long'
                })
                in_position = False
                capital *= 1 + leveraged_return
                returns.append(leveraged_return)
                continue
        elif position_direction == -1:
            trailing_stop = min(trailing_stop, row['close'] * (1 + trailing_stop_pct))
            if row['close'] >= trailing_stop or price_change <= -stop_loss_pct:
                exit_time = row.name
                exit_price = row['close']
                trade_log.append({
                    'entry_time': entry_time,
                    'entry_price': entry_price,
                    'exit_time': exit_time,
                    'exit_price': exit_price,
                    'direction': 'short'
                })
                in_position = False
                capital *= 1 + leveraged_return
                returns.append(leveraged_return)
                continue

    if not in_position:
        if prev_row['macd'] < prev_row['signal'] and row['macd'] > row['signal']:
            in_position = True
            entry_price = row['close']
            position_direction = 1
            trailing_stop = entry_price * (1 - trailing_stop_pct)
            entry_time = row.name
        elif prev_row['macd'] > prev_row['signal'] and row['macd'] < row['signal']:
            in_position = True
            entry_price = row['close']
            position_direction = -1
            trailing_stop = entry_price * (1 + trailing_stop_pct)
            entry_time = row.name

    equity_curve.append(capital)

df['equity'] = pd.Series(equity_curve, index=df.index[:len(equity_curve)]).reindex(df.index, method='ffill')
```

```python
returns = np.array(returns)
cumulative_return = df['equity'].iloc[-1] / initial_capital - 1
win_rate = np.mean(returns > 0)
sharpe_ratio = (np.mean(returns) - risk_free_rate / len(returns)) / (np.std(returns) + 1e-9)
rolling_max = df['equity'].cummax()
drawdown = df['equity'] / rolling_max - 1
max_drawdown = drawdown.min()

print(f"Total Return: {cumulative_return:.2%}")
print(f"Win Rate: {win_rate:.2%}")
print(f"Sharpe Ratio: {sharpe_ratio:.2f}")
print(f"Max Drawdown: {max_drawdown:.2%}")
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df.index[:len(equity_curve)], equity_curve, label='Strategy Equity')
plt.axhline(initial_capital, color='gray', linestyle='--', label='Initial Capital')
plt.title("MACD Strategy with Leverage & Trailing Stop")
plt.legend()
plt.show()

plt.figure(figsize=(12, 4))
plt.plot(drawdown, label='Drawdown')
plt.fill_between(drawdown.index, drawdown, color='red', alpha=0.3)
plt.title("Drawdown Curve")
plt.legend()
plt.show()
```
